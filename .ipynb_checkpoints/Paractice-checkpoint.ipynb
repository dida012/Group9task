```python
# Make sure inline plots are enabled
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Measurement_summary.csv")

# Convert the "Measurement date" column to datetime
df['Measurement date'] = pd.to_datetime(df['Measurement date'])

# Plot PM2.5 over time
plt.figure(figsize=(12,6))
plt.plot(df['Measurement date'], df['PM2.5'], marker='o', linestyle='-', color='blue')
plt.title('PM2.5 Trend Over Time')
plt.xlabel('Measurement Date')
plt.ylabel('PM2.5 (g/m)')
plt.xticks(rotation=45)   # Rotate x-axis labels for readability
plt.grid(True)
plt.tight_layout()        # Adjust layout to prevent overlap
plt.show()
```

```python
region_pm25 = df.groupby('Address')['PM2.5'].mean()
print(region_pm25)
```

```python
import matplotlib.pyplot as plt

# Bar chart of average PM2.5 by region
plt.figure(figsize=(10,6))
region_pm25.plot(kind='bar', color='skyblue')
plt.title('Average PM2.5 by Region')
plt.xlabel('Region')
plt.ylabel('Average PM2.5 (g/m)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Group by region and calculate mean for multiple pollutants
region_pollutants = df.groupby('Address')[['PM2.5','PM10']].mean()
print(region_pollutants)

# Plot side-by-side bar chart
region_pollutants.plot(kind='bar', figsize=(10,6))
plt.title('Average Pollution Levels by Region')
plt.xlabel('Address')
plt.ylabel('Concentration (g/m)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
